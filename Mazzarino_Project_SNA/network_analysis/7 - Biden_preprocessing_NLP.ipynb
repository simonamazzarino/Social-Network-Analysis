```python
import pandas as pd
pd.low_memory = False
```

### Load and clean scraped data

```python
df1 = pd.read_csv("../data_collection/data_biden/primo_semestre.csv",  lineterminator='\n')
df2 = pd.read_csv("../data_collection/data_biden/secondo_semestre.csv",  lineterminator='\n')
df3 = pd.read_csv("../data_collection/data_biden/terzo_semestre.csv",  lineterminator='\n')
df4 = pd.read_csv("../data_collection/data_biden/quarto_semestre.csv",  lineterminator='\n')
df5 =pd.read_csv("../data_collection/data_biden/quinto_semestre.csv",  lineterminator='\n')
```

### Load and process community DF

```python
df_community1 = pd.read_csv("../data_collection/data_biden/Community_Louvain1.csv", na_filter=True, na_values='[]')
df_community2 = pd.read_csv("../data_collection/data_biden/Community_Louvain2.csv", na_filter=True, na_values='[]')
df_community3 = pd.read_csv("../data_collection/data_biden/Community_Louvain3.csv", na_filter=True, na_values='[]')
df_community4 = pd.read_csv("../data_collection/data_biden/Community_Louvain4.csv", na_filter=True, na_values='[]')
df_community5 = pd.read_csv("../data_collection/data_biden/Community_Louvain5.csv", na_filter=True, na_values='[]')
```

```python
dict_comm1 = df_community1.to_dict(orient="list")
dict_comm2 = df_community2.to_dict(orient="list")
dict_comm3 = df_community3.to_dict(orient="list")
dict_comm4 = df_community4.to_dict(orient="list")
dict_comm5 = df_community5.to_dict(orient="list")
```

```python
dict_comm_transformed1 = {user: community for community, users in dict_comm1.items() for user in users}
dict_comm_transformed2 = {user: community for community, users in dict_comm2.items() for user in users}
dict_comm_transformed3 = {user: community for community, users in dict_comm3.items() for user in users}
dict_comm_transformed4 = {user: community for community, users in dict_comm4.items() for user in users}
dict_comm_transformed5 = {user: community for community, users in dict_comm5.items() for user in users}
```

```python
df_community1 = pd.DataFrame.from_dict(dict_comm_transformed1, orient='index',columns=["community"])
df_community2 = pd.DataFrame.from_dict(dict_comm_transformed2, orient='index',columns=["community"])
df_community3 = pd.DataFrame.from_dict(dict_comm_transformed3, orient='index',columns=["community"])
df_community4 = pd.DataFrame.from_dict(dict_comm_transformed4, orient='index',columns=["community"])
df_community5 = pd.DataFrame.from_dict(dict_comm_transformed5, orient='index',columns=["community"])
```

```python
df_community1 = df_community1.rename_axis('username').reset_index()
df_community2 = df_community2.rename_axis('username').reset_index()
df_community3 = df_community3.rename_axis('username').reset_index()
df_community4 = df_community4.rename_axis('username').reset_index()
df_community5 = df_community5.rename_axis('username').reset_index()
```

```python
df_community1['original_community_extension'] = df_community1['community'].map(df_community1['community'].value_counts())
df_community2['original_community_extension'] = df_community2['community'].map(df_community2['community'].value_counts())
df_community3['original_community_extension'] = df_community3['community'].map(df_community3['community'].value_counts())
df_community4['original_community_extension'] = df_community4['community'].map(df_community4['community'].value_counts())
df_community5['original_community_extension'] = df_community5['community'].map(df_community5['community'].value_counts())
```

### Merge on username

```python
df_nlp1 = (pd.merge(df1, df_community1, left_on='username', right_on='username', how='inner').drop('username', axis=1))
df_nlp2 = (pd.merge(df2, df_community2, left_on='username', right_on='username', how='inner').drop('username', axis=1))
df_nlp3 = (pd.merge(df3, df_community3, left_on='username', right_on='username', how='inner').drop('username', axis=1))
df_nlp4 = (pd.merge(df4, df_community4, left_on='username', right_on='username', how='inner').drop('username', axis=1))
df_nlp5 = (pd.merge(df5, df_community5, left_on='username', right_on='username', how='inner').drop('username', axis=1))


```
